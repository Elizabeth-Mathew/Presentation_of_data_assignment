### 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

car_df = pd.read_csv("cars_data.csv")
```

```python
car_df['Buyer Gender'].value_counts().plot(kind='bar', figsize=(5, 6))
plt.xlabel("Gender")
plt.ylabel("Count of People")
plt.title("A bar graph of male vs female buyers participated in the sales");
```

Insights

The count of female buyers participated in the sales is more the male buyers.
